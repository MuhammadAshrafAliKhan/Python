```python
class car:
    def __init__(self, model, speed):
        self.model = model
        self.speed = speed
        self.is_running = False # Spelling sahi kar di
    
    def start(self):
        self.is_running = True
        print(f"The {self.model} is started.")

    def drive(self, value):
        if self.is_running: # Check karein ke car start hai ya nahi
            self.speed += value
            return f"The speed of the car is now {self.speed} km/h"
        else:
            return "Start the car first!"

    def stop(self):
        self.is_running = False
        self.speed = 0
        return f"The {self.model} has stopped."

# Testing
c1 = car("Toyota", 50)
c1.start()

# Ab print use karein taake output dikhe
print(c1.drive(30)) 
print(c1.stop())
```

```python
class AtM:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
        self.is_acount_blocked = False

    def deposit(self, amount):
        if self.is_acount_blocked:
            return "Transaction Failed: Account is Blocked!"
        self.balance += amount
        return f"Successfully deposited {amount}. New Balance: {self.balance}"
    
    # 1. Yahan 'amount' ko argument mein add kiya
    def withdraw(self, amount):
        if self.is_acount_blocked:
            return "Transaction Failed: Account is Blocked!"
        
        # 2. 'self.amount' ki jagah sirf 'amount' use kiya
        if amount > self.balance:
            return "Insufficient balance!"
        
        self.balance -= amount
        return f"Successfully withdrawn {amount}. Remaining Balance: {self.balance}"

# Test karke dekhte hain
a1 = AtM("Ali", 1000)
print(a1.withdraw(500))  # Sahi chalega
print(a1.withdraw(2000)) # "Insufficient balance" dikhayega

    
```

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculate_area(self):
        area = self.length * self.width
        return f"The Area is: {area}"

    def calculate_perimeter(self):
        perimeter = 2 * (self.length + self.width)
        return f"The Perimeter is: {perimeter}"

    def is_square(self):
        # Check karne ke liye ke kya ye rectangle ek square bhi hai
        if self.length == self.width:
            return "Yes, it is also a Square!"
        else:
            return "No, it is a regular Rectangle."

# --- Object Test ---
# Case 1: Square ke liye (Length aur Width barabar)
sq = Rectangle(45, 45)
print("--- Square Test (45x45) ---")
print(sq.is_square())
print(sq.calculate_area())
print(sq.calculate_perimeter())

print("\n")

# Case 2: Rectangle ke liye
rect = Rectangle(10, 5)
print("--- Rectangle Test (10x5) ---")
print(rect.is_square())
print(rect.calculate_area())
print(rect.calculate_perimeter())
```
