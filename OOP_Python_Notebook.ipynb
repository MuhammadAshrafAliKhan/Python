## Instance Variables & Methods

**Instance variables** belong to each object separately. **Instance methods** use `self`.

```python
class Car:
    def __init__(self, model, speed):
        self.model = model
        self.speed = speed
        self.is_running = False

    def start(self):
        self.is_running = True
        print(f"{self.model} started")

    def accelerate(self, value):
        self.speed += value
        print(f"{self.model} speed: {self.speed} km/h")

    def stop(self):
        self.is_running = False
        self.speed = 0
        print(f"{self.model} stopped")

car = Car("Toyota", 0)
car.start()
car.accelerate(60)
car.accelerate(20)
car.stop()
```

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited {amount}. Balance: {self.balance}")

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew {amount}. Balance: {self.balance}")
        else:
            print("Insufficient funds")

acc = BankAccount("Ali", 1000)
acc.deposit(500)
acc.withdraw(200)
acc.withdraw(2000)
```

```python
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius

    def to_fahrenheit(self):
        return (self.celsius * 9/5) + 32

    def to_kelvin(self):
        return self.celsius + 273.15

    def show(self):
        print(f"{self.celsius}°C = {self.to_fahrenheit()}°F = {self.to_kelvin()}K")

t1 = Temperature(0)
t2 = Temperature(100)
t1.show()
t2.show()
```

## 🏦 Project: ATM Machine

```python
class ATM:
    def __init__(self, owner, pin, balance=0):
        self.owner = owner
        self.__pin = pin
        self.balance = balance
        self.transaction_history = []

    def verify_pin(self, entered_pin):
        return entered_pin == self.__pin

    def check_balance(self, pin):
        if self.verify_pin(pin):
            print(f"Balance: PKR {self.balance}")
        else:
            print("Wrong PIN")

    def deposit(self, pin, amount):
        if self.verify_pin(pin):
            self.balance += amount
            self.transaction_history.append(f"Deposited: {amount}")
            print(f"PKR {amount} deposited. New balance: {self.balance}")
        else:
            print("Wrong PIN")

    def withdraw(self, pin, amount):
        if self.verify_pin(pin):
            if amount <= self.balance:
                self.balance -= amount
                self.transaction_history.append(f"Withdrew: {amount}")
                print(f"PKR {amount} withdrawn. Remaining: {self.balance}")
            else:
                print("Insufficient balance")
        else:
            print("Wrong PIN")

    def mini_statement(self, pin):
        if self.verify_pin(pin):
            print(f"--- Mini Statement for {self.owner} ---")
            for t in self.transaction_history:
                print(t)
            print(f"Current Balance: {self.balance}")
        else:
            print("Wrong PIN")

my_atm = ATM("Ali Khan", 1234, 5000)

```

## Class Variables & Class Methods

**Class variables** are shared by all objects. Use `@classmethod` with `cls`.

```python
class Student:
    school_name = "City School"
    total_students = 0

    def __init__(self, name, grade):
        self.name = name
        self.grade = grade
        Student.total_students += 1

    @classmethod
    def get_school(cls):
        return cls.school_name

    @classmethod
    def get_count(cls):
        return cls.total_students

    def info(self):
        print(f"{self.name} | Grade: {self.grade} | School: {Student.school_name}")
```

## Constructors & Destructors

`__init__` runs on object creation. `__del__` runs when object is deleted.

```python
class FileHandler:
    def __init__(self, filename):
        self.filename = filename
        print(f"Opening file: {filename}")

    def read(self):
        print(f"Reading from {self.filename}")

    def write(self, data):
        print(f"Writing '{data}' to {self.filename}")

    def __del__(self):
        print(f"Closing file: {self.filename}")

f = FileHandler("data.txt")
f.read()
f.write("Hello World")
del f
```

### METHOD OVERRIDING — Simple Example

```python
class Employee:
    def __init__(self, name, base_salary):
        self.name = name
        self.base_salary = base_salary

    def calculate_salary(self):
        return self.base_salary

    def display(self):
        print(f"{self.name} — Salary: PKR {self.calculate_salary()}")


class Manager(Employee):
    def __init__(self, name, base_salary, bonus):
        super().__init__(name, base_salary)
        self.bonus = bonus

    def calculate_salary(self):         # OVERRIDES parent's version!
        return self.base_salary + self.bonus


class Developer(Employee):
    def __init__(self, name, base_salary, overtime_hours):
        super().__init__(name, base_salary)
        self.overtime_hours = overtime_hours

    def calculate_salary(self):         # OVERRIDES parent's version!
        overtime_pay = self.overtime_hours * 500
        return self.base_salary + overtime_pay


class Intern(Employee):
    def calculate_salary(self):         # OVERRIDES parent's version!
        return self.base_salary * 0.5   # Intern gets half salary

```

```python

# Testing
e = Employee("Ali", 50000)
m = Manager("Sara", 80000, 20000)
d = Developer("Ahmed", 60000, 40)
i = Intern("Hassan", 30000)

e.display()
m.display()
d.display()
i.display()
```

### POLYMORPHISM 

```python
class Payment:
    def __init__(self, amount):
        self.amount = amount

    def process(self):
        print(f"Processing PKR {self.amount}...")

    def receipt(self):
        print(f"Payment of PKR {self.amount} — {self.process_type()}")


class CashPayment(Payment):
    def process_type(self):
        return "Paid by CASH 💵"


class CardPayment(Payment):
    def __init__(self, amount, card_number):
        super().__init__(amount)
        self.card_number = card_number

    def process_type(self):
        masked = "****" + self.card_number[-4:]
        return f"Paid by CARD 💳 ({masked})"


class OnlinePayment(Payment):
    def __init__(self, amount, app_name):
        super().__init__(amount)
        self.app_name = app_name

    def process_type(self):
        return f"Paid via {self.app_name} 📱"


# ═══════════════════════════════════════
# POLYMORPHISM IN ACTION
# ═══════════════════════════════════════

payments = [
    CashPayment(5000),
    CardPayment(15000, "4532789012345678"),
    OnlinePayment(8000, "JazzCash"),
    CashPayment(2000),
    OnlinePayment(12000, "EasyPaisa")
]

print("═" * 45)
print("         PAYMENT RECEIPTS")
print("═" * 45)

for p in payments:
    p.receipt()          # SAME call for ALL payment types!

print("═" * 45)
print(f"Total Payments: {len(payments)}")
total = sum(p.amount for p in payments)
print(f"Total Amount: PKR {total}")
```

```python
class Doctor:
    total_doctors = 0

    def __init__(self, name, specialty):
        self.name = name
        self.specialty = specialty
        Doctor.total_doctors += 1

    def info(self):
        print(f"Dr. {self.name} — {self.specialty}")

    @classmethod
    def count(cls):
        print(f"Total Doctors: {cls.total_doctors}")


class Patient:
    total_patients = 0

    def __init__(self, name, age, disease):
        self.name = name
        self.age = age
        self.disease = disease
        self.assigned_doctor = None
        Patient.total_patients += 1
        print(f"Patient {name} admitted")

    def assign_doctor(self, doctor):
        self.assigned_doctor = doctor
        print(f"{self.name} assigned to Dr. {doctor.name}")

    def status(self):
        if self.assigned_doctor:
            doc = self.assigned_doctor.name 
        else:"Not Assigned"
        print(f"Patient: {self.name} | Age: {self.age} | Disease: {self.disease} | Doctor: {doc}")

    @classmethod
    def count(cls):
        print(f"Total Patients: {cls.total_patients}")

    def __del__(self):
        print(f"Patient {self.name} discharged")


```

## Single Inheritance

Child class inherits all variables and methods from one parent.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print(f"I am {self.name}, {self.age} years old")

class Employee(Person):
    def __init__(self, name, age, emp_id, salary):
        super().__init__(name, age)
        self.emp_id = emp_id
        self.salary = salary

    def work(self):
        print(f"{self.name} (ID:{self.emp_id}) is working")

    def get_raise(self, percent):
        self.salary += self.salary * percent / 100
        print(f"New salary: {self.salary}")

```

```python
e = Employee("Ali", 28, "E101", 50000)
e.introduce()
e.work()
e.get_raise(10)
```

```python
class Vehicle:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

    def move(self):
        print(f"{self.brand} moving at {self.speed} km/h")

    def info(self):
        print(f"Brand: {self.brand}, Max Speed: {self.speed}")

class Car(Vehicle):
    def __init__(self, brand, speed, doors):
        super().__init__(brand, speed)
        self.doors = doors

    def honk(self):
        print(f"{self.brand}: Beep beep!")

    def info(self):
        super().info()
        print(f"Doors: {self.doors}")

c = Car("Toyota", 180, 4)
c.move()
c.honk()
c.info()
```

## Multilevel Inheritance

A chain: Grandparent → Parent → Child

```python
class Employee:
    def __init__(self, name, emp_id):
        self.name = name
        self.emp_id = emp_id

    def show_basic(self):
        print(f"ID: {self.emp_id} | Name: {self.name}")

class Manager(Employee):
    def __init__(self, name, emp_id, department):
        super().__init__(name, emp_id)
        self.department = department
        self.team = []

    def add_team_member(self, member):
        self.team.append(member)

class Director(Manager):
    def __init__(self, name, emp_id, department, budget):
        super().__init__(name, emp_id, department)
        self.budget = budget

    def show_full(self):
        self.show_basic()
        print(f"Dept: {self.department} | Budget: {self.budget} | Team: {len(self.team)}")

```

```python
d = Director("Khalid", "D01", "Technology", 5000000)
d.add_team_member("Ali")
d.add_team_member("Sara")

d.add_team_member("Umar")
d.show_full()
```

```python
class LivingThing:
    def __init__(self, name):
        self.name = name

    def breathe(self):
        print(f"{self.name} is breathing")

class Animal(LivingThing):
    def __init__(self, name, legs):
        super().__init__(name)
        self.legs = legs

    def move(self):
        print(f"{self.name} moves on {self.legs} legs")

class Dog(Animal):
    def __init__(self, name, breed , legs):
        super().__init__(name, legs)
        self.breed = breed

    def bark(self):
        print(f"{self.name} ({self.breed}): Woof!")

```

```python
d = Dog("Rex", "Labrador", 4)
d.breathe()
d.move()
d.bark()
```

## Hierarchical Inheritance

One parent class, multiple child classes.

```python
class Shape:
    def __init__(self, color="black"):
        self.color = color

    def describe(self):
        print(f"I am a {self.color} shape")

    def area(self):
        return 0

class Circle(Shape):
    def __init__(self, radius, color="blue"):
        super().__init__(color)
        self.radius = radius

    def area(self):
        return 3.14159 * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, width, height, color="red"):
        super().__init__(color)
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

class Triangle(Shape):
    def __init__(self, base, height, color="green"):
        super().__init__(color)
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height

```

```python
shapes = [Circle(5), Rectangle(4, 6), Triangle(3, 8)]
for s in shapes:
    s.describe()
    print(f"Area: {s.area():.2f}")
```

```python
class Employee:
    def __init__(self, name, base_salary):
        self.name = name
        self.base_salary = base_salary

    def info(self):
        print(f"{self.name} — Base Salary: {self.base_salary}")

    def calculate_pay(self):
        return self.base_salary

class FullTime(Employee):
    def __init__(self, name, base_salary, bonus):
        super().__init__(name, base_salary)
        self.bonus = bonus

    def calculate_pay(self):
        return self.base_salary + self.bonus

class PartTime(Employee):
    def __init__(self, name, hourly_rate, hours):
        super().__init__(name, 0)
        self.hourly_rate = hourly_rate
        self.hours = hours

    def calculate_pay(self):
        return self.hourly_rate * self.hours

class Freelancer(Employee):
    def __init__(self, name, project_fee):
        super().__init__(name, 0)
        self.project_fee = project_fee

    def calculate_pay(self):
        return self.project_fee

staff = [FullTime("Ali", 50000, 10000), PartTime("Sara", 500, 80), Freelancer("Umar", 35000)]
for emp in staff:
    print(f"{emp.name}: PKR {emp.calculate_pay()}")
```

## Multiple Inheritance & MRO

A class inherits from more than one parent. MRO defines the lookup order.

```python
class Father:
    def __init__(self):
        self.eye_color = "brown"

    def cooking(self):
        print("Father cooks biryani")

    def trait(self):
        print("Father's trait: hardworking")

class Mother:
    def __init__(self):
        self.hair_color = "black"

    def singing(self):
        print("Mother sings beautifully")

    def trait(self):
        print("Mother's trait: creative")

class Child(Father, Mother):
    def __init__(self):
        Father.__init__(self)
        Mother.__init__(self)

    def show(self):
        print(f"Eye: {self.eye_color}, Hair: {self.hair_color}")

c = Child()
c.cooking()
c.singing()
c.trait()
c.show()
print("MRO:", [cls.__name__ for cls in Child.__mro__])
```

```python
class Flyable:
    def fly(self):
        print(f"{self.name} is flying")

    def move(self):
        print(f"{self.name} moves through air")

class Swimmable:
    def swim(self):
        print(f"{self.name} is swimming")

    def move(self):
        print(f"{self.name} moves through water")

class Duck(Flyable, Swimmable):
    def __init__(self, name):
        self.name = name

    def quack(self):
        print(f"{self.name}: Quack!")

d = Duck("Donald")
d.fly()
d.swim()
d.quack()
d.move()
print("MRO:", [cls.__name__ for cls in Duck.__mro__])
```

```python
class Serializable:
    def serialize(self):
        data = {k: v for k, v in self.__dict__.items()}
        print(f"Serialized: {data}")

class Printable:
    def print_info(self):
        for k, v in self.__dict__.items():
            print(f"  {k}: {v}")

class Loggable:
    def log(self, message):
        print(f"[LOG] {message}")

class Product(Serializable, Printable, Loggable):
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

    def sell(self, qty):
        if qty <= self.stock:
            self.stock -= qty
            self.log(f"Sold {qty} of {self.name}")
        else:
            self.log(f"Not enough stock for {self.name}")

p = Product("Laptop", 85000, 10)
p.print_info()
p.sell(3)
p.serialize()
print("MRO:", [cls.__name__ for cls in Product.__mro__])
```

```python
class Person:
    def __init__(self, name, age, phone):
        self.name = name
        self.age = age
        self.phone = phone

    def contact(self):
        print(f"Contact: {self.name} — {self.phone}")


class Teacher(Person):
    total_teachers = 0

    def __init__(self, name, age, phone, subject, salary):
        super().__init__(name, age, phone)
        self.subject = subject
        self.salary = salary
        Teacher.total_teachers += 1

    def teach(self):
        print(f"Mr/Ms {self.name} is teaching {self.subject}")

    @classmethod
    def count(cls):
        print(f"Total Teachers: {cls.total_teachers}")


class Student(Person):
    total_students = 0

    def __init__(self, name, age, phone, roll_no, grade):
        super().__init__(name, age, phone)
        self.roll_no = roll_no
        self.grade = grade
        self.courses = []
        Student.total_students += 1

    def enroll(self, subject):
        self.courses.append(subject)
        print(f"{self.name} enrolled in {subject}")

    def report(self):
        print(f"--- Student Report ---")
        print(f"Name: {self.name} | Roll: {self.roll_no} | Grade: {self.grade}")
        print(f"Courses: {', '.join(self.courses)}")

    @classmethod
    def count(cls):
        print(f"Total Students: {cls.total_students}")


class Principal(Teacher):
    def __init__(self, name, age, phone, subject, salary, school_name):
        super().__init__(name, age, phone, subject, salary)
        self.school_name = school_name

    def announce(self, msg):
        print(f"📢 Announcement from Principal {self.name}: {msg}")


p = Principal("Dr. Tariq", 55, "0321-0000000", "Admin", 150000, "City School")
t1 = Teacher("Ms. Ayesha", 35, "0311-1111111", "Mathematics", 60000)
t2 = Teacher("Mr. Bilal", 40, "0312-2222222", "Physics", 65000)

s1 = Student("Ali", 16, "0333-1111111", "F101", "A")
s2 = Student("Zara", 15, "0333-2222222", "F102", "A+")

p.announce("Final exams start next Monday!")
t1.teach()
t2.teach()

s1.enroll("Mathematics")
s1.enroll("Physics")
s2.enroll("Mathematics")
s2.enroll("Chemistry")

s1.report()
s2.report()

Teacher.count()
Student.count()
```

## Access Specifiers

Public: `name` | Protected: `_name` | Private: `__name`

```python
class BankAccount:
    bank_name = "National Bank"

    def __init__(self, owner, balance, pin):
        self.owner = owner
        self._balance = balance
        self.__pin = pin

    def show_balance(self, entered_pin):
        if entered_pin == self.__pin:
            print(f"Balance: PKR {self._balance}")
        else:
            print("Incorrect PIN")

    def _calculate_interest(self):
        return self._balance * 0.05

    def show_interest(self):
        print(f"Annual Interest: PKR {self._calculate_interest():.2f}")

class SavingsAccount(BankAccount):
    def __init__(self, owner, balance, pin):
        super().__init__(owner, balance, pin)

    def add_interest(self):
        interest = self._calculate_interest()
        self._balance += interest
        print(f"Interest added: {interest}. New balance: {self._balance}")

acc = SavingsAccount("Ali", 10000, 5678)
acc.show_balance(5678)
acc.add_interest()
acc.show_balance(5678)
```

```python
class Employee:
    def __init__(self, name, department, salary, employee_id):
        self.name = name
        self.department = department
        self._salary = salary
        self.__employee_id = employee_id

    def get_id_last4(self):
        id_str = str(self.__employee_id)
        return "****" + id_str[-4:]

    def _get_salary(self):
        return self._salary

    def public_info(self):
        print(f"Name: {self.name} | Dept: {self.department} | ID: {self.get_id_last4()}")

class Manager(Employee):
    def __init__(self, name, department, salary, employee_id, team_size):
        super().__init__(name, department, salary, employee_id)
        self.team_size = team_size

    def show_salary_range(self):
        print(f"Salary range access via protected: {self._salary} - {self._salary * 1.2:.0f}")

m = Manager("Hamza", "IT", 80000, 10042567, 5)
m.public_info()
m.show_salary_range()
```

```python
class MedicalRecord:
    def __init__(self, patient_name, age, blood_type, diagnosis, insurance_id):
        self.patient_name = patient_name
        self.age = age
        self._blood_type = blood_type
        self.__diagnosis = diagnosis
        self.__insurance_id = insurance_id

    def public_summary(self):
        print(f"Patient: {self.patient_name}, Age: {self.age}")

    def _get_blood_type(self):
        return self._blood_type

    def __get_full_record(self):
        return f"Diagnosis: {self.__diagnosis}, Insurance: {self.__insurance_id}"

    def authorized_view(self, role):
        if role == "doctor":
            print(self.__get_full_record())
        elif role == "nurse":
            print(f"Blood Type: {self._get_blood_type()}")
        else:
            self.public_summary()

rec = MedicalRecord("Fatima", 28, "O+", "Diabetes Type 2", "INS-2024-9871")
rec.authorized_view("visitor")
rec.authorized_view("nurse")
rec.authorized_view("doctor")
```

## Name Mangling

Python renames `__var` to `_ClassName__var` internally.

```python
class Wallet:
    def __init__(self, owner, amount):
        self.owner = owner
        self.__amount = amount

    def show(self):
        print(f"{self.owner}'s wallet: PKR {self.__amount}")

    def add(self, money):
        self.__amount += money

    def spend(self, money):
        if money <= self.__amount:
            self.__amount -= money
            print(f"Spent {money}. Remaining: {self.__amount}")
        else:
            print("Not enough money")

w = Wallet("Ali", 5000)
w.show()
w.add(2000)
w.spend(1500)
print(dir(w))
print(w._Wallet__amount)
```

```python
class Base:
    def __init__(self):
        self.__secret = "I am base secret"

    def show_secret(self):
        print(self.__secret)

class Child(Base):
    def __init__(self):
        super().__init__()
        self.__secret = "I am child secret"

    def show_child_secret(self):
        print(self.__secret)

b = Base()
c = Child()
b.show_secret()
c.show_secret()
c.show_child_secret()
print(c._Base__secret)
print(c._Child__secret)
```

```python
class Config:
    def __init__(self, db_host, db_pass, api_key):
        self.db_host = db_host
        self.__db_pass = db_pass
        self.__api_key = api_key

    def connect(self, password):
        if password == self.__db_pass:
            print(f"Connected to {self.db_host}")
            return True
        print("Connection failed: wrong password")
        return False

    def get_api(self, password):
        if password == self.__db_pass:
            return self.__api_key
        return "Access Denied"

cfg = Config("localhost", "admin123", "XK-2024-SECRET")
cfg.connect("wrong")
cfg.connect("admin123")
print(cfg.get_api("admin123"))
print(cfg._Config__api_key)
```

## Inner / Nested Classes

A class defined inside another class.

```python
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
        self.engine = Car.Engine(1600, "Petrol")

    def show(self):
        print(f"{self.brand} {self.model}")
        self.engine.show()

    class Engine:
        def __init__(self, cc, fuel):
            self.cc = cc
            self.fuel = fuel

        def show(self):
            print(f"  Engine: {self.cc}cc | Fuel: {self.fuel}")

        def start(self):
            print("  Engine started: Vroom!")

car = Car("Honda", "City")
car.show()
car.engine.start()
```

```python
class University:
    def __init__(self, name, city):
        self.name = name
        self.city = city
        self.departments = []

    def add_department(self, dept_name, head):
        dept = University.Department(dept_name, head)
        self.departments.append(dept)
        return dept

    def show(self):
        print(f"{self.name} — {self.city}")
        for d in self.departments:
            d.show()

    class Department:
        def __init__(self, name, head):
            self.name = name
            self.head = head
            self.courses = []

        def add_course(self, course):
            self.courses.append(course)

        def show(self):
            print(f"  Dept: {self.name} | Head: {self.head} | Courses: {', '.join(self.courses)}")

uni = University("FAST University", "Lahore")
cs = uni.add_department("Computer Science", "Dr. Hassan")
math = uni.add_department("Mathematics", "Prof. Nazia")
cs.add_course("OOP")
cs.add_course("DSA")
math.add_course("Calculus")
uni.show()
```

```python
class Laptop:
    def __init__(self, brand, ram):
        self.brand = brand
        self.ram = ram
        self.battery = Laptop.Battery(5000)
        self.keyboard = Laptop.Keyboard("Mechanical")

    def power_on(self):
        print(f"{self.brand} powering on...")
        self.battery.check()
        self.keyboard.ready()

    class Battery:
        def __init__(self, mah):
            self.mah = mah
            self.level = 100

        def check(self):
            print(f"  Battery: {self.level}% ({self.mah}mAh)")

        def drain(self, percent):
            self.level = max(0, self.level - percent)

    class Keyboard:
        def __init__(self, ktype):
            self.ktype = ktype

        def ready(self):
            print(f"  Keyboard ready: {self.ktype}")

l = Laptop("Dell", "16GB")
l.power_on()
l.battery.drain(30)
l.battery.check()
```

## Association, Aggregation & Composition

Three ways classes relate: uses-a, has-a(weak), has-a(strong).

```python
class Doctor:
    def __init__(self, name, specialization):
        self.name = name
        self.specialization = specialization

class Patient:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Appointment:
    def __init__(self, doctor, patient, date):
        self.doctor = doctor
        self.patient = patient
        self.date = date

    def show(self):
        print(f"Appointment: {self.patient.name} → Dr.{self.doctor.name} on {self.date}")

doc = Doctor("Ali", "Cardiology")
pat = Patient("Zara", 30)
apt = Appointment(doc, pat, "2024-12-10")
apt.show()
```

```python
class Teacher:
    def __init__(self, name, subject):
        self.name = name
        self.subject = subject

    def info(self):
        print(f"Teacher: {self.name} | Subject: {self.subject}")

class Department:
    def __init__(self, name):
        self.name = name
        self.teachers = []

    def add_teacher(self, teacher):
        self.teachers.append(teacher)

    def show(self):
        print(f"Department: {self.name}")
        for t in self.teachers:
            t.info()

t1 = Teacher("Ms. Ayesha", "Math")
t2 = Teacher("Mr. Bilal", "Physics")

dept = Department("Science")
dept.add_teacher(t1)
dept.add_teacher(t2)
dept.show()

del dept
t1.info()
```

```python
class Room:
    def __init__(self, room_type, size):
        self.room_type = room_type
        self.size = size
        print(f"Room created: {room_type}")

    def info(self):
        print(f"  {self.room_type} — {self.size} sqft")

    def __del__(self):
        print(f"Room destroyed: {self.room_type}")

class House:
    def __init__(self, address, room_data):
        self.address = address
        self.rooms = [Room(rt, sz) for rt, sz in room_data]

    def show(self):
        print(f"House: {self.address}")
        for r in self.rooms:
            r.info()

    def __del__(self):
        print(f"House at {self.address} demolished")

h = House("Street 5, Lahore", [("Bedroom", 200), ("Kitchen", 150), ("Lounge", 300)])
h.show()
print("Deleting house...")
del h
```

## 🛒 Project: E-Commerce Order System

```python
class Address:
    def __init__(self, street, city, country):
        self.street = street
        self.city = city
        self.country = country

    def full(self):
        return f"{self.street}, {self.city}, {self.country}"


class Product:
    _total_products = 0

    def __init__(self, name, price, stock):
        self.name = name
        self._price = price
        self._stock = stock
        self.__product_id = f"P{Product._total_products + 1:04d}"
        Product._total_products += 1

    def is_available(self, qty):
        return self._stock >= qty

    def reduce_stock(self, qty):
        self._stock -= qty

    def info(self):
        print(f"[{self.__product_id}] {self.name} — PKR {self._price} | Stock: {self._stock}")

    @classmethod
    def catalog_count(cls):
        print(f"Total products in catalog: {cls._total_products}")


class Customer:
    def __init__(self, name, email, address):
        self.name = name
        self.__email = email
        self.address = address
        self.order_history = []

    def get_email_masked(self):
        parts = self.__email.split("@")
        return parts[0][:2] + "***@" + parts[1]

    def add_order(self, order_id):
        self.order_history.append(order_id)

    def show_profile(self):
        print(f"Customer: {self.name}")
        print(f"Email: {self.get_email_masked()}")
        print(f"Delivery: {self.address.full()}")
        print(f"Orders placed: {len(self.order_history)}")


class OrderItem:
    def __init__(self, product, quantity):
        self.product = product
        self.quantity = quantity
        self.subtotal = product._price * quantity

    def show(self):
        print(f"  {self.product.name} x{self.quantity} = PKR {self.subtotal}")


class Order:
    _order_count = 0

    def __init__(self, customer):
        Order._order_count += 1
        self.__order_id = f"ORD-{Order._order_count:04d}"
        self.customer = customer
        self.items = []
        self.status = "Pending"
        print(f"Order {self.__order_id} created for {customer.name}")

    def add_item(self, product, qty):
        if product.is_available(qty):
            product.reduce_stock(qty)
            item = OrderItem(product, qty)
            self.items.append(item)
            print(f"Added: {product.name} x{qty}")
        else:
            print(f"Sorry, {product.name} has insufficient stock")

    def total(self):
        return sum(i.subtotal for i in self.items)

    def confirm(self):
        self.status = "Confirmed"
        self.customer.add_order(self.__order_id)
        print(f"Order {self.__order_id} CONFIRMED!")

    def invoice(self):
        print(f"{'='*40}")
        print(f"INVOICE — {self.__order_id}")
        print(f"Customer: {self.customer.name}")
        print(f"Deliver to: {self.customer.address.full()}")
        print(f"{'='*40}")
        for item in self.items:
            item.show()
        print(f"{'='*40}")
        print(f"TOTAL: PKR {self.total()}")
        print(f"Status: {self.status}")
        print(f"{'='*40}")


addr = Address("Street 5, Block B", "Lahore", "Pakistan")
customer = Customer("Ali Hassan", "ali.hassan@email.com", addr)

p1 = Product("Laptop", 85000, 5)
p2 = Product("Mouse", 1500, 20)
p3 = Product("USB Hub", 2200, 8)
p4 = Product("Laptop Bag", 3500, 3)

Product.catalog_count()

order = Order(customer)
order.add_item(p1, 1)
order.add_item(p2, 2)
order.add_item(p3, 1)
order.add_item(p4, 5)
order.confirm()
order.invoice()
customer.show_profile()
```
